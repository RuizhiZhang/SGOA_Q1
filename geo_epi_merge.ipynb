```python
import pandas as pd
```

```python
epi = pd.read_excel('data/JHUData.xlsx')
epi = epi.drop('Unnamed: 0', 1)
```

```python
geo = pd.read_excel('题目一：地名数据库.xlsx', sheet_name='霍普金斯地名')
geo = geo.drop('Column1', 1)
```

```python
merge = pd.merge(geo,epi)
merge
```

```python
tmp = merge[merge['Province_State'].isna()]

for i in tmp.index:
    merge['Province_State'].iloc[i] = 'Unassigned'
```

```python
merge.sort_values(by='Country_Region').reset_index().drop('index',1).to_excel('data/geo_epi.xlsx')
```
