```python
import os
import pandas as pd

df = pd.DataFrame(columns = ['Province/State', 'Country/Region', 'Last Update', 'Confirmed', 'Deaths', 'Recovered', 'Latitude', 'Longitude'])
directory = '/Users/ruizhi/Documents/国网大数据中心 题目/题目一/data/COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports'
for root,dirs,files in os.walk(directory):
    for file in files:
        df = df.append(pd.read_csv(directory + '/' + file))
```

```python
df
```

```python
dt = df
```

```python
dt.info()
```

```python
mask = dt['Last_Update'].isna()
a = dt[mask]
a = a.rename(columns={'Last Update': 'last_update'})

mask = dt['Last Update'].isna()
b = dt[mask]
b = b.rename(columns={'Last_Update': 'last_update'})

dt = a.append(b).drop(['Last_Update', 'Last Update'], 1)
```

```python
mask = dt['Province_State'].isna()
a = dt[mask]
a = a.rename(columns={'Province/State': 'province_state'})

mask = dt['Province/State'].isna()
b = dt[mask]
b = b.rename(columns={'Province_State': 'province_state'})

dt = a.append(b).drop(['Province_State', 'Province/State'], 1)
```

```python
mask = dt['Country_Region'].isna()
a = dt[mask]
a = a.rename(columns={'Country/Region': 'country_region'})

mask = dt['Country/Region'].isna()
b = dt[mask]
b = b.rename(columns={'Country_Region': 'country_region'})

dt = a.append(b).drop(['Country_Region', 'Country/Region'], 1)
```

```python
mask = dt['Latitude'].isna()
a = dt[mask]
a = a.rename(columns={'Lat': 'latitude'})

mask = dt['Lat'].isna()
b = dt[mask]
b = b.rename(columns={'Latitude': 'latitude'})

dt = a.append(b).drop(['Latitude', 'Lat'], 1)
```

```python
mask = dt['Longitude'].isna()
a = dt[mask]
a = a.rename(columns={'Long_': 'longitude'})

mask = dt['Long_'].isna()
b = dt[mask]
b = b.rename(columns={'Longitude': 'longitude'})

dt = a.append(b).drop(['Longitude', 'Long_'], 1)
```

```python
dt = dt.drop(['.DS_Store','Unnamed: 0'], 1)
```

```python
dt['last_update'] = pd.to_datetime(dt['last_update'])
```

```python
dt['last_update'] = dt['last_update'].dt.strftime('%m/%d/%y %H:%M')
```

```python
dt
```

```python
dt = dt.replace('Mainland China', 'China')
```

```python
dt = dt[~((dt['country_region'] == 'US') & (dt['FIPS'].isna()))]
```

```python
dt = dt.sort_values(by=['FIPS','province_state','country_region','last_update'])
```

```python
final = pd.DataFrame(columns = ['index','FIPS', 'Admin2', 'province_state', 'country_region', 'last_update', 'longitude', 'latitude','Confirmed', 'Deaths', 'Recovered', 'Active', 'Combined_Key'])
mask = ~dt['FIPS'].isna()

tmp = dt[mask].reset_index().drop('index', 1)
for i in tmp.index:
    if i == len(tmp) - 1:
        final = final.append(tmp.iloc[i])
    if i < len(tmp) - 1 and (tmp['FIPS'].iloc[i] != tmp['FIPS'].iloc[i + 1]):
        final = final.append(tmp.iloc[i])
```

```python
final
```

```python
mask = (dt['FIPS'].isna() & ~dt['province_state'].isna())
tmp = dt[mask].reset_index().drop('index', 1)
for i in tmp.index:
    if i == len(tmp) - 1:
        final = final.append(tmp.iloc[i])
    if i < len(tmp) - 1 and (tmp['province_state'].iloc[i] != tmp['province_state'].iloc[i + 1]):
        final = final.append(tmp.iloc[i])
    
mask = (dt['FIPS'].isna() & dt['province_state'].isna())
tmp = dt[mask].reset_index().drop('index', 1)
for i in tmp.index:
    if i == len(tmp) - 1:
        final = final.append(tmp.iloc[i])
    if i < len(tmp) - 1 and (tmp['country_region'].iloc[i] != tmp['country_region'].iloc[i + 1]):
        final = final.append(tmp.iloc[i])
```

```python
final
```

```python
final = final.drop('index', 1).rename(columns={'province_state': 'Province_State', 
                                       'country_region':'Country_Region',
                                       'last_update':'Last_Update',
                                       'longitude':'Longitude',
                                       'latitude':'Latitude'})
```

```python
final.sort_values(by=['Country_Region','Province_State']).reset_index().drop('index',1).to_excel('data/JHUData.xlsx')
```
